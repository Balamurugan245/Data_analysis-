# Chennai Super Kings Team Analysis 

```python
CSK = df[df['TEAM'] == 'Chennai Super Kings']
fig, axes = plt.subplots(1,3, figsize=(30,10))

sns.barplot(x=CSK["PLAYER"],y=CSK["PRICE PAID"],color="yellow",ax=axes[0])
axes[0].set_title("Chennai Super Kings Players Price")


CSK_NATIONALITY = CSK.groupby('NATIONALITY')['PLAYER'].count()
sns.barplot(x=CSK_NATIONALITY.index, y=CSK_NATIONALITY,color="yellow",estimator=sum,ax=axes[1])
axes[1].set_xlabel("NATIONALITY")
axes[1].set_ylabel("PLAYERS NATIONALITY COUNT")
axes[1].set_title("CSK TEAM PLAYERS NATIONALITY")

CSK_TYPE= CSK.groupby('TYPE')['PLAYER'].count()
sns.barplot(x=CSK_TYPE.index, y=CSK_TYPE,color="yellow",estimator=sum,ax=axes[2])
axes[2].set_xlabel("TYPE")
axes[2].set_ylabel("PLAYERS TYPE COUNT")
axes[2].set_title("CSK TEAM PLAYERS TYPE COUNT")

plt.tight_layout()
plt.show()
```

```python
CSK_players = CSK.groupby('TYPE')['PRICE PAID'].sum()
fig, axes = plt.subplots(1, 2, figsize=(30, 10))

custom_colors = ['gold', 'lightskyblue']

axes[0].pie(CSK_players, labels=CSK_players.index, autopct='%1.1f%%',colors=custom_colors)
axes[0].set_title('Chennai Super Kings Players Type Price')

CSK_NATIONALITY = CSK.groupby('NATIONALITY')['PRICE PAID'].sum()
axes[1].pie(CSK_NATIONALITY, labels=CSK_NATIONALITY.index, autopct='%1.1f%%',colors=custom_colors)
axes[1].set_title('Chennai Super Kings Players Nationality Price')

plt.show()
```
